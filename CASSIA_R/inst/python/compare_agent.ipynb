```python
def filter_and_compare(csv_path, 
                      correctness_level=None,  # 0.5, 1, or 0
                      score_range=(0, 100),    # tuple of (min_score, max_score)
                      tissue=None,             # specific tissue to filter
                      dataset=None):           # specific dataset to filter
    
    # Read the CSV file
    df = pd.read_csv(csv_path)
    
    # Apply filters
    mask = pd.Series([True] * len(df))
    
    if correctness_level is not None:
        mask &= df['Correctness'] == correctness_level
    
    if score_range is not None:
        mask &= df['Score'].between(score_range[0], score_range[1])
        
    if tissue is not None:
        mask &= df['Tissue'] == tissue
        
    if dataset is not None:
        mask &= df['dataset'] == dataset
    
    filtered_df = df[mask]
    
    # Process each filtered row
    results = {}  # Changed to dictionary with cell type as key
    for idx, row in filtered_df.iterrows():
        # Extract marker list and format it
        markers = row['Marker List'].split(',')
        marker_set = "\n".join([f"{i+1}. {marker.strip()}" for i, marker in enumerate(markers)])
        
        # Get the first subtype from Predicted.Sub.Cell.Types
        subtypes = row['Predicted.Sub.Cell.Types'].split(',')
        first_subtype = subtypes[0].strip()
        
        # Create a unique key using cell type and tissue
        result_key = f"{row['True Cell Type']}_{row['Tissue']}"
        
        # Get the comparison results
        comparison = compare_models(
            tissue=row['Tissue'],
            celltypeA=row['True Cell Type'],
            celltypeB=first_subtype,
            marker_set=marker_set
        )
        
        results[result_key] = {
            'dataset': row['dataset'],
            'tissue': row['Tissue'],
            'true_cell_type': row['True Cell Type'],
            'predicted_main_type': row['Predicted.Main.Cell.Type'],
            'predicted_subtype': first_subtype,
            'marker_set': marker_set,
            'model_responses': comparison
        }
    
    return results

def print_results(results):
    """Helper function to print results in a organized way"""
    for cell_type_key, result in results.items():
        print(f"\n=== Results for {cell_type_key} ===")
        print(f"Dataset: {result['dataset']}")
        print(f"Tissue: {result['tissue']}")
        print(f"True Cell Type: {result['true_cell_type']}")
        print(f"Predicted Main Type: {result['predicted_main_type']}")
        print(f"Predicted Subtype Used: {result['predicted_subtype']}")
        print("\nMarker Set:")
        print(result['marker_set'])
        print("\nModel Responses:")
        for model, response in result['model_responses'].items():
            print(f"\n--- {model} ---")
            print(response)
        print("\n" + "="*50)

# Example usage:
def run_example():
    results = filter_and_compare(
        csv_path="your_data.csv",
        correctness_level=0.5,
        score_range=(95, 100),
        tissue="Kidney",
        dataset="Azimuth"
    )
    print_results(results)
```

```python
def create_html_report(results_dict):
    html = """
    <!DOCTYPE html>
    <html>
    <head>
        <style>
            body { font-family: Arial, sans-serif; margin: 20px; }
            .cell-type-section { 
                border: 1px solid #ddd; 
                margin: 20px 0; 
                padding: 20px;
                border-radius: 5px;
            }
            .model-response { 
                background-color: #f9f9f9; 
                padding: 15px; 
                margin: 10px 0;
                border-left: 4px solid #2196F3;
            }
            .metadata { 
                display: grid;
                grid-template-columns: repeat(2, 1fr);
                gap: 10px;
                margin-bottom: 15px;
            }
            .metadata-item {
                background-color: #f0f0f0;
                padding: 8px;
                border-radius: 4px;
            }
            .marker-set {
                background-color: #e8f5e9;
                padding: 10px;
                border-radius: 4px;
                margin: 10px 0;
            }
            h2 { color: #333; }
            h3 { color: #666; }
        </style>
    </head>
    <body>
        <h1>Cell Type Comparison Results</h1>
    """
    
    for cell_type_key, result in results_dict.items():
        html += (
            f'<div class="cell-type-section">'
            f'<h2>{cell_type_key}</h2>'
            '<div class="metadata">'
            f'<div class="metadata-item"><strong>Dataset:</strong> {result["dataset"]}</div>'
            f'<div class="metadata-item"><strong>Tissue:</strong> {result["tissue"]}</div>'
            f'<div class="metadata-item"><strong>True Cell Type:</strong> {result["true_cell_type"]}</div>'
            f'<div class="metadata-item"><strong>Predicted Main Type:</strong> {result["predicted_main_type"]}</div>'
            f'<div class="metadata-item"><strong>Predicted Subtype:</strong> {result["predicted_subtype"]}</div>'
            '</div>'
            '<div class="marker-set">'
            '<h3>Marker Set:</h3>'
            f'<pre>{result["marker_set"]}</pre>'
            '</div>'
            '<h3>Model Responses:</h3>'
        )
        
        for model, response in result['model_responses'].items():
            html += (
                '<div class="model-response">'
                f'<h4>{model}</h4>'
                f'<p>{response.replace(chr(10), "<br>")}</p>'
                '</div>'
            )
        
        html += "</div>"
    
    html += """
    </body>
    </html>
    """
    
    return html

def save_html_report(results_dict, output_path):
    """Save the HTML report to a file"""
    html_content = create_html_report(results_dict)
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(html_content)
    print(f"Report saved to: {output_path}")

# Example usage:
save_html_report(hcl_results, "cell_type_comparison_report.html")
```
